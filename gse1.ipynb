```
import pandas as pd
import plotly.express as px
```

**Analyse GSE**

```
tx_long_central = pd.read_excel('VARECO_TIRAGES_TXLONG_Central_VA.xlsx', header=1, index_col=0)
tx_long_txMoins100 = pd.read_excel('VARECO_TIRAGES_TXLONG_Tx-100.xlsx', header=1, index_col=0)
tx_long_txPlus100 = pd.read_excel('VARECO_TIRAGES_TXLONG_Tx_Spd+100.xlsx', header=1, index_col=0)
tx_long_txPlus200 = pd.read_excel('VARECO_TIRAGES_TXLONG_Tx_Spd+200.xlsx', header=1, index_col=0)
```

```
tx_inf_central = pd.read_excel('VARECO_TIRAGES_TXINF_Central_VA.xlsx', header=1, index_col=0)
```

```
pd.DataFrame({'taux-100bp': tx_long_txMoins100.mean(), 'central':tx_long_central.mean(), 'taux+100bp':tx_long_txPlus100.mean(),'taux+200bp':tx_long_txPlus200.mean()})
```

```
moyennes = pd.DataFrame({'taux-100bp': tx_long_txMoins100.mean(), 'central':tx_long_central.mean(), 'taux+100bp':tx_long_txPlus100.mean(),'taux+200bp':tx_long_txPlus200.mean()})
fig = px.line(moyennes)
fig.update_layout(
    yaxis_tickformat = '0.00%',
        title="Moyennes",
    xaxis_title="mois de projection",
    yaxis_title="taux long",
   
  
    )
fig.show()
```

```

fig = px.line(tx_long_txPlus200.describe().transpose().drop(['count', 'std'], axis=1))
fig.update_layout(
    yaxis_tickformat = '0.00%',
        title="taux+200bp",
    xaxis_title="mois de projection",
    yaxis_title="taux long",
   
  
    )
fig.show()
```

```
# remplacer mean par max, min, etc. si besoin. 
fig = px.box(tx_long.mean(axis=1) , points="all")
fig.update_layout(
    yaxis_tickformat = '0.00%',
        title="",
    xaxis_title="taux moyen",
    yaxis_title="",
  
    )
fig.show()
```

```
tx_long_all = pd.DataFrame(tx_long.values.flatten(), columns=['taux long'])
fig = px.box(tx_long_all)
fig.update_layout(
    yaxis_tickformat = '0.00%',
        title="",
    xaxis_title="taux",
    yaxis_title="",
  
    )
fig.show()
```

```

fig = px.histogram(tx_long_all)
fig.update_layout(
    xaxis_tickformat = '0.00%',
        title="",
    xaxis_title="taux",
    yaxis_title="",
  
    )
fig.show()
```

```
sc_max = tx_long[tx_long.max(axis=1)>=0.5]

#fig = px.line(sc_max.transpose().values.tolist())
fig = px.line(sc_max.transpose())
fig.update_layout(
    yaxis_tickformat = '0.00%',
        title="Scénarios où le plafond 50% est atteint",
    xaxis_title="mois de projection",
    yaxis_title="taux long",
   
  
    )
fig.show()
```

**Analyse BE**

```
BE = pd.read_excel('EV09.xlsx', header=5, index_col=0).iloc[:,0]
BE.describe()
```

```
fig = px.histogram(BE)
fig.update_layout(
    
    title="Histogramme BE par scénario",
    xaxis_title="BE (en K€)",
    yaxis_title="Nb. d'occurence",
  
    )
fig.show()
```

```
fig = px.box(BE)
fig.update_layout(
  
        title="",
    xaxis_title="BE (en K€)",
    yaxis_title="",
  
    )
fig.show()
```

```
BE_taux_max = BE.loc[sc_max.index]
fig = px.histogram(BE_taux_max)
fig.update_layout(
    
    title="Histogramme BE par scénario",
    xaxis_title="BE (en K€)",
    yaxis_title="Nb. d'occurence",
  
    )
fig.show()

fig = px.box(BE_taux_max)
fig.update_layout(
  
        title="",
    xaxis_title="BE (en K€)",
    yaxis_title="",
  
    )
fig.show()
```

```
BE_extremes = BE[(BE>29000000) & (BE<30000000)]


sc_BE_extremes = tx_long.loc[BE_extremes.index]

#fig = px.line(sc_max.transpose().values.tolist())
fig = px.line(sc_BE_extremes.transpose())
fig.update_layout(
    yaxis_tickformat = '0.00%', 
    title="Scénarios dont BE entre 29 Md€ et 30 M€",
    xaxis_title="mois de projection",
    yaxis_title="taux long",
   
  
    )
fig.show()
```
